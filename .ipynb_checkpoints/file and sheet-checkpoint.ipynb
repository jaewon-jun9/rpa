```python
#모든 하위폴더 내의 엑셀파일에서 워크시트의 이름 가져오기
import openpyxl
for i in xlsxs:
    load_wb = openpyxl.load_workbook(i, data_only=True) # data_only=True 를 선택할 경우 수식이 아닌 값을 가져옴
    print (i,load_wb.sheetnames) #문서 내의 시트 이름 보기
```

```python
#특정 폴더 내에서 엑셀파일의 목록을 가져온 후 모든 파일에서 두 번째 워크시트의 내용을 출력한다.
#폴더를 지정하고
setFolder='D:/2020project/하남시/LCC 0907'
# 하위폴더를 포함하여 모든 엑셀 파일을 리스트에 담기
import glob
targetPattern1 = setFolder+"/*/*.xlsx"
# targetPattern2 = "./*.xls"
xlsxs=glob.glob(targetPattern1)
# xlss=glob.glob(targetPattern2)
# xlsxs=xlsxs+xlss
print(xlsxs)
#모든 하위폴더 내의 엑셀파일에서 워크시트의 이름 가져오기
import openpyxl
for i in xlsxs:
    load_wb = openpyxl.load_workbook(i, data_only=True) # data_only=True 를 선택할 경우 수식이 아닌 값을 가져옴
    print (i,load_wb.sheetnames) #문서 내의 시트 이름 보기
    load_ws = load_wb[load_wb.sheetnames[1]] #'두 번째에 위치한 시트 지정
    all_values=[] # 시트의 내용을 답을 리스트
    for row in load_ws.rows: #row 단위로 읽기
        row_value=[] #row 마다 셀의 값을 저장
        for cell in row: #row 마다 cell 단위로 데이터 읽기
            row_value.append(cell.value) #row_value에 cell의 내용을 넣어주어 row를 완성
        all_values.append(row_value) #row를 전체 리스트에 추가    
    print(all_values)
```

```python
#특정 폴더 내에서 엑셀파일의 목록을 가져온 후 모든 파일에서 두 번째 워크시트의 내용을 출력한다.
#폴더를 지정하고
setFolder='D:/2020project/하남시/LCC 0907'
# 하위폴더를 포함하여 모든 엑셀 파일을 리스트에 담기
import glob
targetPattern1 = setFolder+"/*/*.xlsx"
# targetPattern2 = "./*.xls"
xlsxs=glob.glob(targetPattern1)
# xlss=glob.glob(targetPattern2)
# xlsxs=xlsxs+xlss
print(xlsxs)
#모든 하위폴더 내의 엑셀파일에서 워크시트의 이름 가져오기
import openpyxl
sheet2=[]
for i in xlsxs:
    load_wb = openpyxl.load_workbook(i, data_only=True) # data_only=True 를 선택할 경우 수식이 아닌 값을 가져옴
    print (i,load_wb.sheetnames) #문서 내의 시트 이름 보기
    load_ws = load_wb[load_wb.sheetnames[1]] #'두 번째에 위치한 시트 지정
    all_values=[] # 시트의 내용을 답을 리스트
    for row in range(1,11): #row 단위로 읽기
        row_value=[] #row 마다 셀의 값을 저장
        for cell in range(1,11): #row 마다 cell 단위로 데이터 읽기
            row_value.append(load_ws.cell(row,cell).value) #row_value에 cell의 내용을 넣어주어 row를 완성
        all_values.append(row_value) #row를 전체 리스트에 추가    
    print(all_values)
```
